```
data=pd.concat([df1,df2,df4,df5,df6,df26])
#print(data['failure'])
```

```
import matplotlib as plt
import numpy as np
print(len(data))
data.hist(bins=10,figsize=(50,50))
```

```
#from sklearn.linear_model import Lasso
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(data['serial_number'])
#print(le.classes_)
data['serial_number']=le.transform(data['serial_number'])
le.fit(data['model'])
#print(le.classes_)
data['model']=le.transform(data['model'])
le.fit(data['date'])
#print(le.classes_)
data['date']=le.transform(data['date'])
```

```
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

print("\n Initial feature \n")
print(data)
all_feature_names=data.columns
print(all_feature_names)
features_count=0
for feature in all_feature_names:
  features_count+=1
  print(str(features_count)+""+feature)
```

```
import numpy as np
y= np.array(data['capacity_bytes']).reshape(len(data),1)
x=data.drop(['capacity_bytes'],axis=1)
x=x.values
print(type(x))
print(x)
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(x)
x=imp.fit_transform(x)
print(x)
```

```
import matplotlib.pyplot as plt
alpha_value=1
while (alpha_value<=8):
  clf = Lasso(alpha = alpha_value)
  sfm = SelectFromModel(clf)
  sfm.fit(x, y)
  n_features = sfm.transform(x).shape[1]
  print(n_features)
  objects=('Before LASSO','After LASSO')
  count=[features_count,n_features]
  plt.bar(objects,count,align='center')
  plt.ylabel('features_count')
  plt.title('Feature selection using LASSO(alpha= '+str(alpha_value)+')')
  plt.show()
  print('selected features when Alpha= '+str(alpha_value)+' are \n')
  selected_feature_indices=sfm.get_support(indices=True)
  count=1
  for index in selected_feature_indices:
    print(str(count)+" "+all_feature_names[index])
    count+=1
  alpha_value+=3
```
