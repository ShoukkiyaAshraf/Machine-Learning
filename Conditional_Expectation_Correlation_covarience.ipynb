# Positive Correlation

```
import numpy as np
np.random.seed(1)
x=np.random.randint(0,50,1000)
y=x+np.random.normal(0,10,1000)
np.corrcoef(x,y)
```

```
import matplotlib 
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
plt.scatter(x,y)
plt.show()
```

```
import numpy as np
np.random.seed(1)
x=np.random.randint(0,50,1000)
y=100-x+np.random.normal(0,5,1000)
np.corrcoef(x,y)
```

```
plt.scatter(x,y)
plt.show()
```

# Correlation Matrix

```
import pandas as pd
df=pd.DataFrame({'a':np.random.randint(0,50,1000)})
df['b']=df['a']+np.random.randint(0,10,1000)
df['c']=100-df['a']+np.random.randint(0,5,1000)
df['d']=np.random.randint(0,50,1000)
df.corr()
```

```

pd.scatter_matrix(df,figsize=(6,6))

plt.show()
```

# Plotting the correlation matrix

```
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
```

# Finding the correlation between many variables (Multidimensional Dataset)

```
data = pd.read_csv('https://www.dropbox.com/s/4jgheggd1dak5pw/data_visualization.csv?raw=1',index_col=1)
```

```
corr_val=data.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax=ax.matshow(corr_val,cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(cax)
ticks = np.arange(0,len(data.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(data.columns)
ax.set_yticklabels(data.columns)
plt.show()
```

**About  Covariance and Correlation**

The variance of a random variable A is $$var(A)=E[(A−E[A])^2]$$, where E[A] is the expected value of A. 
The covariance of A and B is $$Cov(A,B)=E[(A−E[A])(B−E[B])]$$
The correlation of A and B is the covariance of A and B normalized by their variances. That is, $$Corr(A,B)=Cov(A,B)/\sqrt{var(A)var(B)}$$
